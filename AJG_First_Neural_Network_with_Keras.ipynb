```
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```
# Cargar el Dataset
dataset = loadtxt('Diabetes2020.txt', delimiter=',')
#Split Data dentro input (x) y output (y) Variables
x = dataset[:,0:8]
y = dataset[:,8]
```

```
x.shape
```

```
# Definir Modelo de DL de Keras
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```
# Compile the Keras Model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
# Fit el Keras Model en el Dataset
model.fit(x, y, epochs=300, batch_size=10)
```

```
# Evaluar el Modelo Keras
_, accuracy = model.evaluate(x, y)
print('Accuracy : %.2f' % (accuracy*100))
```

```
# Hacer Predicciones sobre Data nunca vista por el Modelo
predictions = model.predict_classes(x)
# Summarize the first 5 Cases
for i in range(5):
  print('%s => %d (esperado %d)' % (x[i].tolist(), predictions[i], y[i]))
```
